# 2. Clustering Coefficient

```python
G=nx.Graph()
G.add_edges_from([(0,1),(1,2),(2,3),(0,2)])
nx.draw(G,with_labels=True)
```

```python
list(G.neighbors(2))
```

node 2 has 3 neighbors. (1,3,0)


```python
print(G.has_edge(1,3))
print(G.has_edge(1,0))
print(G.has_edge(3,0))
```

node 1 and 0 are neighbor each other.

So node 0, 1, and 2 forms a cluster (geometrically, this is triangle)

Maximum number of cluster at node 2 is 3, but there is only 1 cluster.

So clustering coefficient at node 2 is 1/3

```python
nx.clustering(G)
```

```python
nx.average_clustering(G)
```

# 3. Shortest path

```python
np.random.seed(0)
G=nx.gnp_random_graph(10,0.2)
nx.draw(G,with_labels=True)
```

```python
nx.has_path(G,0,5)
```

```python
nx.has_path(G,0,2)
```

```python
nx.shortest_path(G,0,5)
```

```python
nx.shortest_path_length(G,0,5)
```

```python
G=nx.gnp_random_graph(16,0.4)
path_lengths=[]
for i in G.nodes:
  for j in G.nodes:
    if i>j:
      path_lengths.append(nx.shortest_path_length(G,i,j))
path_lengths
```

```python
np.max(path_lengths) # maximum shortest path length
```

```python
nx.diameter(G)
```

```python
np.mean(path_lengths)
```

```python
nx.average_shortest_path_length(G)
```

**Exercise 4. Calculate the average shortest path length of the 3 networks(G1, G2, G3).**

hint: Randomly choose n number of pair of nodes, and then calculate average shortest path length. If you calculate path length for every pair of nodes, it will take lots of time.

```python
def APL(graph,ensemble=100):
  length_list=[]
  nodes=list(graph.nodes)
  for i in range(ensemble):
    u,v=np.random.choice(nodes,size=2,replace=False)
    length_list.append(nx.shortest_path_length(graph,u,v))
  return np.mean(length_list)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = []
with open('HW3.txt') as file:
    for line in file:
        line = line.strip().split()
        mass, height = float(line[0]), float(line[1])
        data.append((mass, height))

mass, height = zip(*data)
```

```python
plt.scatter(mass, height)
plt.xlabel('Mass')
plt.ylabel('Height')
```
